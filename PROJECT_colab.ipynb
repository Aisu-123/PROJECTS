```python
def convert(x):
  try:
    return float(x)
  except:
    return 0
```

The all columns in the dataframe is the datatype of object.we can convert the object data type to float by using a function.

```python
df1['Quantity']=df1['Qty'].apply(convert)
```

By applying the convert function we changed the datatype of the column 'Qty' to float and made a new column named 'Quantity'.

```python
item_qty=df1.groupby("ItemName").sum().unstack()["Quantity"]
item_qty.head()
item_qty.plot(kind="bar",stacked=True)
plt.show()
```

By groupby, we can combine the 'ItemName' and the sum of 'Quantity' and stored it in a variable called 'item_qty'.
By plotting bar plot, we can see that the comparisons among the discrete categories.one axis of the plot shows the ItemName to be compared,and the other axis represents a measured value.

```python
date_sales=df1.groupby("InvDate").count().unstack()["Taxable"].reset_index()
date_sales.head()
date_sales.plot(kind="line",stacked=True)
plt.show()
```

By groupby, we can combine the 'InvDate' and the count of 'Taxable' and stored it in a variable called 'date_sales'.
By plotting line plot, we can see the relationship between two variables.the x- axis of the plot represents the time and the y-axis represents representing the data value.

```python
df1['Amount']=df1['iNetAmt'].apply(convert)
```

By applying the convert function we changed the datatype of the column 'iNetAmt' to float and made a new column named 'Amount'.

```python
item_Amount=df1.groupby("ItemName").sum().unstack()["Amount"]
item_Amount.head()
item_Amount.plot(kind="bar",stacked=True)
plt.show()
```

By groupby, we can combine the 'ItemName' and the sum of 'Amount' and stored it in a variable called 'item_Amount'.
By plotting bar plot, we can see that the comparisons among the discrete categories.one axis of the plot shows the ItemName to be compared,and the other axis represents a measured value.
